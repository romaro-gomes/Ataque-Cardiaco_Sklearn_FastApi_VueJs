### Testando hipoteses

O box-plot é uma maneira rápida de visualizar se existe diferenças entre grupos. Mas para uma confirmação mais acurada é necessáio realizar teste estatisticos para avaliar se a hipotese está ou não correta.

#### Normalidade dos dados
Os testes de Normalidade verificam se as variaveis seguem ou não uma distribuição normal. Isto será um norteado se usaremos testes pareado, para varaivies que segem distribuição normal, ou não pareados, cuja distribuiçao não é normal .

```python
numericas
```

#### Relação entre variaveis categoricas e númericas

```python
for num in numericas:
    valores = dados[num]
    resultado = sp.stats.normaltest(valores) 
    pvalor = resultado.pvalue 
    print(f"O teste de normalidade para a variavel {num} obteve um valor p={pvalor:.2f}. Então {'não é uma distribuição normal' if pvalor < 0.05 else 'é uma distribuição normal'}") 
```

Como nenhum das varaiaveis segue uma distribuição normal, será utilizado o um teste não-parametrico.

Os testes não paramétricos comparam as medianas entres grupos, pois estas são mais tolerantes a outlers.

Os testes utilizados dependerão do número de grupos/categorias que serão comparadas:
- Mann-Whiteney para comparar duas categorias (p<0.05, medianas diferentes)
- Kruskall-Wallis para comparar mais acima de duas((p<0.05, medianas diferentes)).



```python
def teste_de_mann_whiteney(*,categorias,numericas):
    scale=len(dados[categorias].unique())    
    for colname in numericas:
        list_genre=[]
        for i in range(scale):
            subset=dados[(dados[categorias]==dados[categorias].unique()[i])][colname]
            list_genre.append(subset.tolist())
                
        res = sp.stats.mstats.mannwhitneyu(*list_genre)
        print(f'O teste de Mann-Whitney quando compara se há diferença entre as categorias de {categorias} em relação há {colname} obtém o p-valor de {res.pvalue:.3f}.\nIsto indica que {"há diferença" if res.pvalue < 0.05 else "não há diferença"} entre as categorias de {categorias}.')
        print('')
        
```

```python
teste_de_mann_whiteney(categorias='diagnostico',numericas=numericas)
```

#### Correlação entre variaveis númericas

Visto que existe uma diferençã significativa entre as medianas dos variaveis númericas de pessoas com diagnóstico positivo e negativo,torna-se necessário avalia sealums delas se relacionam entre si.

A principal forma de se verificar é através de testes de correlação. Não indicados utilizar variaveis altamente relacionados na construção de modelos de regressão, sendo uma parte importante na seleção de variaveis/características/features.

Após selecionar o teste adquados, a correalação pode ser avaliada comparando os números, ou de forma visual utilizando uma matriz de correlação, e de forma ainda mais elegante na forma de mapa de calor.

```python
matriz_correlacao = dados[numericas].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(matriz_correlacao, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Mapa de Calor - Correlação entre Variáveis')
plt.show()
```

Todas as correlações estão entre 0.4 e -0.3, o que sugere qye elas sejam independes entre si.

Curiosmente, a idade é inversamente proporcioal a frequência cardiaca, algo muito largamente descrito na literatura.

#### Correlação entre as variaveis categoricas

Também existem testes de hipoteses que estudam se existe alguma relação entre variavesi categoricas, que não seja explicada pelo acaso.

O principal teste para avariguar esta hipotese é o teste de chi-quadrado.

```python
def teste_chi_quadrado(*,categoria,categoricas):
    for cat in categoricas:
        tbContigencia = pd.crosstab(dados[cat], dados[categoria]) 
        print(tbContigencia,'\n')
        resultado = sp.stats.chi2_contingency(tbContigencia) 
        print(f"A relação da entre as categorias {categoria} e {cat} obteve um valor p={resultado.pvalue:.3f}, sendo assim {'há independencia entre as variaveis' if resultado.pvalue>=0.05 else 'as variaveis são relacionadas'}\n") 
        print('*******************************************************************************')
```

```python
teste_chi_quadrado(categoria='diagnostico',categoricas=categoricas)
```
